```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
base = pd.read_excel("State Sumaries.xlsx")
x_axis = base['State']
y_axis = base['Mean']

plt.bar(x_axis, y_axis)

plt.xlabel('Stages')
plt.ylabel('Time (seconds)')
plt.title('Mean Values of Total Time in Each Stage')
plt.legend(loc="upper right")
plt.show()
```

```python
x_axis = base['State']
y_axis = base['Entries']

plt.bar(x_axis, y_axis)

plt.xlabel('Stages')
plt.ylabel('Stages')
plt.title('Total Entries in Each Stage')
plt.show()
```
