```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

## Model 1 Results

```python
case1 = pd.read_excel("./Model 1/30kmh.xlsx")
distance1 = case1['Total distance (meters)']
time1 = case1['Total Time (seconds)']

plt.figure(figsize=(10,5))
plt.plot(time1, label='Total Travel Time (seconds)')
plt.plot(distance1, label='Total Travel Distance (meters)')

plt.xlabel('Number of Agents')
plt.title('Travel Time and Travel Distance for each Agent at 30km/h')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
case2 = pd.read_excel("./Model 1/50kmh.xlsx")
distance2 = case2['Total distance (meters)']
time2 = case2['Total Time (seconds)']

plt.figure(figsize=(10,5))
plt.plot(time2, label='Total Travel Time (seconds)')
plt.plot(distance2, label='Total Travel Distance (meters)')

plt.xlabel('Number of Agents')
plt.title('Travel Time and Travel Distance for each Agent at 50km/h')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
case3 = pd.read_excel("./Model 1/100km.xlsx")
distance3 = case3['Total distance (meters)']
time3 = case3['Total Time (seconds)']

plt.figure(figsize=(10,5))
plt.plot(time3, label='Total Travel Time (seconds)')
plt.plot(distance3, label='Total Travel Distance (meters)')

plt.xlabel('Number of Agents')
plt.title('Travel Time and Travel Distance for each Agent at 100km/h')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
base = pd.read_excel("./Model 1/Total Time.xlsx")
time_base = base['Base']
test1 = base['Test Case 1 (30 km/h)']
test2 = base['Test Case 2 (50 km/h)']
test3 = base['Test Case 3 (100 km/h)']

plt.figure(figsize=(10,5))
plt.plot(time_base, label='Base')
plt.plot(test1, label='Test Case 1 (30 km/h)')
plt.plot(test2, label='Test Case 2 (50 km/h)')
plt.plot(test3, label='Test Case 3 (100 km/h)')

plt.xlabel('Number of Agents')
plt.title('Total Travel Time Comparison between Base and Test Cases')
plt.legend(loc="upper left")
plt.show()
```

```python
avg_distance = np.mean(distance1) * 0.001
print("Average distance between start and end station:", avg_distance, "km")
```

## Model 2 Results

```python
case1 = pd.read_excel("./Model 2/30kmh.xlsx")
distance1 = case1['Total meters']
time1 = case1['Total seconds']
num = case1['Num. Agent']

case2 = pd.read_excel("./Model 2/50kmh.xlsx")
distance2 = case2['Total meters']
time2 = case2['Total seconds']

case3 = pd.read_excel("./Model 2/100kmh.xlsx")
distance3 = case3['Total meters']
time3 = case3['Total seconds']
```

```python
plt.figure(figsize=(10,5))
plt.plot(time1, label='30km/h')
plt.plot(time2, label='50km/h')
plt.plot(time3, label='100km/h')

plt.xlabel('Number of Agents')
plt.title('Travel Time Comparison between different velocity')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
avg_distance1 = np.average(distance1) * 0.001
avg_distance2 = np.average(distance2) * 0.001
avg_distance3 = np.average(distance3) * 0.001

dict = {'30km/h': avg_distance1, '50km/h': avg_distance2, '100km/h': avg_distance3}
cases =list(dict.keys())
avg = list(dict.values())

plt.bar(cases, avg, width=0.5)

plt.xlabel('Test Cases')
plt.ylabel('Travel Distance (km)')
plt.title('Travel Distance Comparison between different velocity')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
