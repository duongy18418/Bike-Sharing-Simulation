```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
case1 = pd.read_excel("150 model - 2.277 speed.xlsx")
distance1 = case1['Total distance (meters)']
time1 = case1['Total Time (seconds)']

plt.figure(figsize=(10,5))
plt.plot(time1, label='Total Travel Time (seconds)')
plt.plot(distance1, label='Total Travel Distance (meters)')

plt.xlabel('Number of Agents')
plt.title('Travel Time and Travel Distance for each Agent')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
base = pd.read_excel("Total Time.xlsx")
time_base = base['Base']
test1 = base['Test Case 1 (100 km/h)']
test2 = base['Test Case 2 (30 km/h)']
test3 = base['Test Case 3 (50 km/h)']

plt.figure(figsize=(10,5))
plt.plot(time_base, label='Base')
plt.plot(test1, label='Test Case 1 (100 km/h)')
plt.plot(test2, label='Test Case 2 (30 km/h)')
plt.plot(test3, label='Test Case 3 (50 km/h)')

plt.xlabel('Number of Agents')
plt.title('Total Travel Time Comparison between Base and Test Cases')
plt.legend(loc="upper left")
plt.show()
```
